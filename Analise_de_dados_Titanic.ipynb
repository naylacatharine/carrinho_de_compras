```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Planilha original
path = "/train_and_test2.csv"
df = pd.read_csv(path)

df.head(5)
```

```python
#Alterei o nome "2urvived" para "surv"
df.rename(columns = {'2urvived':'surv'}, inplace = True)
df.head()
```

```python
age_min = df["Age"].min()
age_max = df["Age"].max()
age_width  = 10
age_bins = nm.arange(age_min , age_max + age_width, age_width)
plt.hist(df["Age"] , bins = age_bins)
plt.xlabel("Idade")
plt.ylabel("Frequencia")
plt.title("Histograma da Idade (Intervalo de 10 em 10)")
plt.show()
```
